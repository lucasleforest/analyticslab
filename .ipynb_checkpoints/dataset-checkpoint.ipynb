```python
#Nb of keys
print(len(dict_map.keys()))

#Rename the market and submarket

def remapping(market, submarket):
    key = (market, submarket)
    return dict_map[key][0], dict_map[key][1]

market_spend.Market, market_spend.Submarket = zip(*market_spend.apply(lambda x : remapping(x.Market, x.Submarket), axis = 1))
market_spend.rename(columns = {"Submarket" : "SubMarket"}, inplace = True)
market_spend
```

```python
market_spend = market_spend[(market_spend.Region == "Worldwide")]
market_spend.drop("Region", axis = 1, inplace = True)
market_spend
```

```python
merged_dataset = cleaned_data.merge(market_spend, how = 'left', on = ['Market', "SubMarket", "Device", "Year"])
merged_dataset = merged_dataset.groupby(["CompanyName", "Device", "Market", "SubMarket", "Year", "Value"]).sum().reset_index()
merged_dataset.rename(columns = {"Market size" : "Spendings"}, inplace = True)
merged_dataset
```

```python
spend_bydevice = merged_dataset[["Device", "Spendings"]].groupby("Device").sum().reset_index().rename(columns = {"Spendings" : "DevSpends"})
spend_bymarket= merged_dataset[["Market", "Spendings"]].groupby("Market").sum().reset_index().rename(columns = {"Spendings" : "MarketSpends"})
spend_bysubmarket = merged_dataset[["SubMarket", "Spendings"]].groupby("SubMarket").sum().reset_index().rename(columns = {"Spendings" : "SubMarketSpends"})
```

```python
spend_bydevice
```

```python
merged_dataset = merged_dataset.merge(spend_bydevice, how = 'left', on = "Device")
merged_dataset = merged_dataset.merge(spend_bymarket, how = 'left', on = "Market")
merged_dataset = merged_dataset.merge(spend_bysubmarket, how = 'left', on = "SubMarket")
merged_dataset
```
