```python
import pandas as pd
import numpy as np
import sklearn
import xgboost
```

```python
df = pd.read_csv("DATA/final_df.csv", index_col=0)
```

```python
df.head()
```

```python
df.rename({"Value":"CompanySpend", "Spendings":"TotalSpend"}, axis=1, inplace=True)
```

```python
df= pd.get_dummies(df, columns=["Device", "Market", "SubMarket"])
```

```python
df.head()
```

```python
X_train = df[(df.Year<=2022)]
X_test = df[(df.Year>=2023)]
y_train = X_train.pop("CompanySpend")
y_test = X_test.pop("CompanySpend")
X_train.drop(["Company", "Year"], axis=1, inplace=True)
X_test.drop(["Company", "Year"], axis=1, inplace=True)
```

```python
import xgboost as xgb
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
model_xgb = xgb.XGBRegressor(verbosity=2, device="cuda")
```

```python

parameters = {'learning_rate': np.arange(0.001, 0.1, 0.005), 
              'max_depth': [4, 5, 6, 7, 8],
              'min_child_weight': [1, 2, 3, 4, 5],
              'subsample': [0.7],
              'colsample_bytree': [0.7],
              'n_estimators': [100, 200, 300, 500],
              'alpha': np.arange(0, 2, 0.1),
              'lambda' : np.arange(0,2,0.1)}

xgb_grid = RandomizedSearchCV(model_xgb,
                        parameters,
                        cv = 4,
                        n_jobs = 1,
                        verbose=3,
                        n_iter = 500)
```

```python
xgb_grid.fit(X_train,
         y_train)
```

```python
best_params = xgb_grid.best_params_
best_score = xgb_grid.best_score_
```

```python
best_params
```

```python
best_score
```

```python
new_xgb = xgb_grid.best_estimator_
```

```python
from sklearn.metrics import r2_score
r2_score(new_xgb.predict(X_test), y_test)
```

```python
from scipy.spatial.distance import pdist
d = pdist(X_train_scaled)

# Perform hierarchical clustering using the Ward variance minimization algorithm
hclust_mod = linkage(d, method='ward')

# Plot the dendrogram
plt.figure(figsize=(10, 7))
dendrogram(hclust_mod, labels=None)
plt.ylabel("Dissimilarity")
plt.xlabel("")
plt.title("")
plt.show()

# Create a DataFrame for dissimilarity values
hc_dissim = pd.DataFrame({
    'k': np.arange(1, len(hclust_mod) + 1),
    'dissimilarity': np.flipud(hclust_mod[:, 2])
})
```

```python
hc_dissim = pd.DataFrame({
    'k': np.arange(1, len(hclust_mod) + 1),
    'dissimilarity': np.flipud(hclust_mod[:, 2])
})
```

```python
plt.figure(figsize=(10, 6))
plt.plot(hc_dissim['k'], hc_dissim['dissimilarity'], marker='o')
plt.title('Scree Plot')
plt.xlabel('Number of Clusters (k)')
plt.xlim(0,100)
plt.ylabel('Dissimilarity')
plt.grid(True)
plt.show()
```

```python
from scipy.cluster.hierarchy import fcluster
clusters = fcluster(hclust_mod, 8, criterion='maxclust')
```
