### Preprocessing

```python
import pandas as pd
import numpy as np
import sklearn
import xgboost
```

```python
df = pd.read_csv("../../clean_data/final_df.csv", index_col=0)
```

```python
df.rename({"Value":"CompanySpend", "Spendings":"TotalSpend"}, axis=1, inplace=True)
df= pd.get_dummies(df, columns=["Device", "Market", "SubMarket"])
```

```python
X_train = df[(df.Year<=2022)]
X_test = df[(df.Year>=2023)]
y_train = X_train.pop("CompanySpend")
y_test = X_test.pop("CompanySpend")
X_train.drop(["Company", "Year"], axis=1, inplace=True)
X_test.drop(["Company", "Year"], axis=1, inplace=True)
```

## Clustering
### "artificial" clustering

```python
#standardize data
from sklearn.preprocessing import MinMaxScaler

# X_train.drop(["Region"], axis = 1, inplace = True) #why was region still there?
# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Fit the scaler to the data and transform it
X_train_normalized = scaler.fit_transform(X_train)

# Convert the array back to a DataFrame (optional, for better readability)
X_train_normalized = pd.DataFrame(X_train_normalized, columns=X_train.columns)

y_train_reshaped = y_train.values.reshape(-1, 1)

# Fit the scaler to the data and transform it
y_train_normalized = scaler.fit_transform(y_train_reshaped)

# Flatten the array back to 1D (if needed)
y_train_normalized = y_train_normalized.ravel()
```

```python
#create clusters based only on "CompanySpend"
import numpy as np
import pandas as pd
from scipy.cluster.hierarchy import linkage, fcluster
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

Z = linkage(y_train_normalized.reshape(-1, 1), method='ward')
clusters = fcluster(Z, t=8, criterion='maxclust')

# Add clusters to X_train
#X_train['cluster'] = clusters
#y_train["cluster"] = clusters
```

```python
#Train a tree on X_train to classify the clusters built above (why is region still in the df??)
tree_clusters = DecisionTreeClassifier(max_depth=6, min_samples_leaf=10)
tree_clusters.fit(X_train, clusters)
```

```python
#Visualize tree and performance metrics
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
from sklearn.metrics import confusion_matrix

# Visualize the decision tree
plt.figure(figsize=(20,10))
plot_tree(tree_clusters, filled=True, feature_names=X_train.columns, class_names=str(np.unique(clusters)))
plt.show()
clusters_pred = tree_clusters.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score
import seaborn as sns

accuracy = accuracy_score(clusters, clusters_pred)
cm = confusion_matrix(clusters, clusters_pred)

plt.figure(figsize=(10,7))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix")
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.show()

print(accuracy)
```

```python
#return feature importance in the tree
tree = tree_clusters.tree_
# Get the feature names
feature_names = X_train.columns

# Find the features used for splitting
split_feature_indices = tree.feature

# Translate indices to feature names, excluding leaves (where feature index is -2)
split_features = [feature_names[i] if i >= 0 else "leaf" for i in split_feature_indices]

# Filter out the "leaf" entries and get unique feature names
split_feature_names = list(set([name for name in split_features if name != "leaf"]))

print(split_feature_names)
```

### Weighted clustering

```python
import sklearn
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from scipy.spatial.distance import pdist
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, confusion_matrix, silhouette_samples, silhouette_score
import seaborn as sns
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv("../../clean_data/final_df.csv", index_col=0)
```

```python
df= pd.get_dummies(df, columns=["Device", "Market", "SubMarket"])
df.rename({"Value":"CompanySpend", "Spendings":"TotalSpend"}, axis=1, inplace=True)
```

```python
scaler = MinMaxScaler()
df_train = df[(df.Year<=2022)]
df_train.drop(["Company", "Year"], axis=1, inplace=True)
df_train_normalized = scaler.fit_transform(df_train)
df_train_normalized = pd.DataFrame(df_train_normalized, columns = df_train.columns)
df_train_normalized_weighted = pd.DataFrame(df_train_normalized, columns = df_train.columns)
df_train_normalized_weighted['CompanySpend'] = df_train_normalized_weighted['CompanySpend'] * 10
```

```python
distance_matrix = pdist(df_train_normalized_weighted, metric='euclidean')
h_clust = linkage(distance_matrix, method='ward')
```

```python
dendro = dendrogram(h_clust)
```

```python
hc_dissim = pd.DataFrame({'k': np.arange(1, len(h_clust) + 1),'dissimilarity': np.flipud(h_clust[:, 2])})

# Create the scree plot
plt.figure(figsize=(10, 6))
plt.plot(hc_dissim['k'], hc_dissim['dissimilarity'], marker='o')
plt.title('Scree Plot')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Dissimilarity')
plt.xlim(0,20)
plt.grid(True)
plt.show()
```

```python
clusters = fcluster(h_clust, t=10, criterion='maxclust')
```

```python
silhouette_vals = silhouette_samples(df_train_normalized_weighted, clusters)
```

```python
# Create a subplot with 1 row and 1 column
fig, ax1 = plt.subplots(1, 1)
fig.set_size_inches(18, 7)

# The 1st subplot is the silhouette plot
# The silhouette coefficient can range from -1, 1 
ax1.set_xlim([-0.1, 1])
ax1.set_ylim([0, len(df) + (10 + 1) * 10])

y_lower = 10
for i in range(10):
    # Aggregate the silhouette scores for samples belonging to cluster i, and sort them
    ith_cluster_silhouette_values = silhouette_vals[clusters == i]
    ith_cluster_silhouette_values.sort()

    size_cluster_i = ith_cluster_silhouette_values.shape[0]
    y_upper = y_lower + size_cluster_i

    color = plt.cm.nipy_spectral(float(i) / 10)
    ax1.fill_betweenx(np.arange(y_lower, y_upper),
                      0, ith_cluster_silhouette_values,
                      facecolor=color, edgecolor=color, alpha=0.7)

    # Label the silhouette plots with their cluster numbers at the middle
    ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

    # Compute the new y_lower for next plot
    y_lower = y_upper + 10  # 10 for the 0 samples

ax1.set_title("The silhouette plot for the various clusters.")
ax1.set_xlabel("The silhouette coefficient values")
ax1.set_ylabel("Cluster label")

# The vertical line for average silhouette score of all the values
silhouette_avg = silhouette_score(df_train_normalized_weighted, clusters)
ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

ax1.set_yticks([])  # Clear the yaxis labels / ticks
ax1.set_xticks([-0.1, 0, 0.2, 0.4, 0.6, 0.8, 1])

# Display the plot
plt.suptitle(("Silhouette analysis for KMeans clustering on sample data "
              "with n_clusters = %d" % 10),
             fontsize=14, fontweight='bold')

plt.show()
```

```python
tree_clusters = DecisionTreeClassifier(max_depth=6, min_samples_leaf=10)
tree_clusters.fit(df_train, clusters)
```
